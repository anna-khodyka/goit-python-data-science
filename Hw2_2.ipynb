Используя метод groupby выполните группировку по столбцу "Должность"

Создайте новый DataFrame, где для сгруппированных данных по столбцу "Должность", выполните агрегацию данных с помощью метода agg и найдите минимальное и максимальное значение в столбце "Зарплата.в.месяц"

```python
grouping_columns=['Должность']
columns_to_show=["Зарплата.в.месяц"]
grouped_df=python_data.groupby(grouping_columns)[columns_to_show].agg([np.min, np.max])
grouped_df
```

Проверочная таблица для следующего задания:

```python
grouping_columns=['Должность']
columns_to_show=["Зарплата.в.месяц"]
grouped_df=python_data.groupby(grouping_columns)[columns_to_show].agg([np.mean, np.min, np.max])
grouped_df=grouped_df.sort_values(by=('Зарплата.в.месяц', 'mean'))

# рисую столбиковую диаграмму
grouped_df.plot(kind='bar', label=['average', 'min', 'max'])
plt.legend()
plt.show()
```

Создайте функцию fill_avg_salary которая будет возвращать среднее значение заработной платы в месяц. Используйте ее для метода apply и создайте новый столбик "avg"

```python
def fill_avg_salary(row):
    # возвращает среднюю зарплату в месяц для Должности
    return python_data[python_data['Должность']==row['Должность']]["Зарплата.в.месяц"].mean()

# создаем столбец с Средними зарплатами
fill_avg_series=python_data.apply(fill_avg_salary, axis=1)
# добавляем созданный столбец в DataFrame:
python_data.insert(loc=len(python_data.columns), column='Средняя.зарплата', value=fill_avg_series)
python_data
```

```python
# устанавливаю порядок в 'Уровень.английского'
pd.Categorical(python_data['Уровень.английского'], categories=['элементарный', 'ниже среднего', 'средний', 'выше среднего', 'продвинутый'], ordered=True)

# устанавливаю размер полотна для диаграммы
plt.figure(figsize=(8, 4))

# рисуем swarmplot диаграмму
sns.swarmplot(x='Уровень.английского', y='salary', hue='Должность', data=python_data)
plt.title('Зарплата в зависимости от уровня английского языка')
plt.xlabel('Уровень английского')
plt.ylabel('Зарплата')
plt.show()
```

```python
# scatterplot общий опыт / зп
# преобразовываем Общий.опыт.работы в float
# python_data['Общий.опыт.работы'].replace(to_replace="10 и более лет", value=10, inplace=True)
# python_data['Общий.опыт.работы'].replace(to_replace="меньше 3 месяцев", value=0.25, inplace=True)
python_data['Общий.опыт.работы']=python_data['Общий.опыт.работы'].astype('float64')

# устанавливаю размер полотна для диаграммы
plt.figure(figsize=(8, 4))

# рисуем scatterplot диаграмму
sns.scatterplot(x='Общий.опыт.работы', y='salary', data=python_data)
plt.title('Зарплата в зависимости от опыта работы')
plt.xlabel('Опыт работы, лет')
plt.ylabel('Зарплата')
plt.show()
```
