```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
parsing=pd.read_html("https://goit.global/textbooks/python/data-science-xf8b2g/v1/docs/module-03/hw", match="Момент наблюдения")
stocks=parsing[0]
print(stocks.head())
```

Постройте линии регрессии, с помощью функции regplot, зависимости доходности одной акции от другой. Всего три графика. Выберите пару акций которые наименее зависимы в доходностях от друг друга.

```python
sns.regplot(x="Доходность A, %", y="Доходность B, %", data=stocks, label = "A vs. B")
sns.regplot(x="Доходность A, %", y="Доходность C, %", data=stocks, label = "A vs. C")
sns.regplot(x="Доходность B, %", y="Доходность C, %", data=stocks, label = "B vs. C")
plt.legend()
plt.show()
```

```python
sns.regplot(x="Доходность A, %", y="Доходность B, %", data=stocks)
```

```python
sns.regplot(x="Доходность A, %", y="Доходность C, %", data=stocks)
```

```python
sns.regplot(x="Доходность B, %", y="Доходность C, %", data=stocks)
```

ОТВЕТ: наименее зависимые акции - А и В: линия регресии зависимости идет практически паралельно оси, доверительный интервал широкий.
Как подтверждение: кореляция составляет 0,07.
